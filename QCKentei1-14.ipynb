# Garantía de calidad

## 14-01 Garantía de calidad

### 1. Garantía de resultados y garantía por proceso
*  No se trata solo de garantizar los resultados mediante la inspección, sino de centrarse en el proceso (el trabajo) y gestionarlo y asegurar la calidad el mismo. Esto también se expresa como que la calidad se construye en el proceso.
* Variabilidad del sistema de factores

### 2. Garantía y compensación
* Garantía: acciones enfocadas antes de que ocurra el problema.
* Compensación: acciones posteriores, principalmente de índole económica, después de que ocurra el problema

### 3. Diagrama del sistema de garantía de calidad
* Es la asignación de las tareas relacionadas con la garantía de calidad en cada paso, desde el deseó y desarrollo del producto, la fabricación, la inspección, el envío, la venta, el servicio postventa, hasta el manejo de quejas. El sistema de garantía de calidad se muestra generalmente en un diagrama de flujo.
* Conexión de los procesos
Criterios de evaluación y organismos de reuniones
* Normas y registros
* Roles de cada departamento. Ubicación de la responsabilidad

### 4. Despliegue de la función de calidad (QFD)
* QFD: Quality Function Deployment. Es una metodología que utiliza diversas transformaciones y despliegues para lograr los objetivos de calidad del producto.
* Tabla binaria
* Tabla de calidad: Requisitos de calidad del cliente, Características de calidad del producto

### 5. DR y predicción de problemas FMEA FTA

### 6. Proceso de garantía de calidad y red de garantía
* Es un diagrama que resume en qué proceso se deben realizar la prevención de ocurrencia y la  prevención de salida, a partir de la correspondencia entre los defectos y los procesos.
* Grado de garantía / Estado de las medidas
* Prevención de ocurrencia / Prevención de salida
* Proceso donde se deben tomar acciones
* Comprensión común
* Habilidades / Transmisión de conocimientos

### 7. Garantía de calidad a lo largo de todo el ciclo de vida del producto
* Es la idea de garantizar la calidad no solo en el momento de la venta del producto, sino también durante su uso e incluso hasta su eliminación.
* LCA: Life Cycle Assessment
* Método para evaluar cuantitativamente la carga ambiental

### 8. Seguridad del producto, consideraciones ambientales y responsabilidad por productos
* La seguridad del producto se refiere a la creación de productos seguros para garantizar la seguridad del cliente durante el uso del producto.
* La consideración ambiental también se requiere en las normas relacionadas con las especificaciones del producto, así como en los métodos de prueba y evaluación correspondientes.
* La responsabilidad por productos defectuosos (PL: Product Liability) es la responsabilidad de indemnización que debe asumir el fabricante por los daños personales o materiales causados por un defecto en el producto.
* Defecto de diseño
* Defecto de fabricación
* Defecto en las instrucciones o advertencia
* Prevención de responsabilidad por productos defectuosos (PLP: Product Liability Prevention)
* Defensa contra la responsabilidad por productos defectuosos (PLD: Product Liability Defence)

### 9. Gestión de la fluidez inicial
* Se trata de realizar una gestión especial, diferenciada de los productos anteriores, para la detección temprana y tratamiento rápido de los problemas relacionados con los defectos de calidad que no se detectaron en las etapas de diseño y prototipo, los cuales se hacen evidentes al inicio de la producción de un nuevo producto o tras cambios significativos en el proceso de fabricación, con el objetivo de estabilizar la calidad lo antes posible.

### 10. Manejo de problemas en el mercado, quejas y su resolución
* Una queja es una expresión de insatisfacción hacía una organización respecto a un producto, servicio o el proceso de gestión de quejas, en la que se espera explícita o implícitamente una respuesta o resolución.



## 14-02 Garantía del proceso

### 1. Instrucción de trabajo (IT)
* Documento de estándar de trabajo: Documento escrito del estándar de estándar de trabajo
*Estándar de trabajo: Estándar que indica el propósito del trabajo, las condiciones de trabajo (materiales utilizados, equipos y herramientas, entorno de trabajo, etc.), el método de trabajo (incluida la seguridad) y el método de verificación de los resultados del trabajo (autocontrol de calidad y cantidad, entre otros).

### 2. Concepto de proceso (o etapa del proceso)
* Es un conjunto de actividades interrelacionadas o que interactúan entre sí para convertir una entrada en una salida.

### 3. Diagrama de proceso QC y Diagrama de flujo
* Es una representación en diagrama de flujo del proceso desde los materiales y componentes del producto hasta su envío como producto terminado, en la que se clarifican los elementos de control y los métodos de control en cada etapa del proceso.

### 4. Concepto de anormalidades en el proceso, su detección y acciones correctivas
* Que el proceso no esté en estado de control y que deje de estar en estado estable debido a causas no despreciables.

### 5. Investigación de la capacidad del proceso y análisis del proceso
* La capacidad del proceso se refiere a la capacidad del proceso en relación con la calidad. La investigación de la capacidad del proceso consiste en tomar muestras del proceso, medir las características de calidad y estimar la capacidad del proceso.
* El análisis del proceso es una actividad que busca esclarecer la relación entre las características y los factores a partir de los datos recopilados en el proceso. A menudo se utilizan métodos estadísticos para mantener y mejorar el proceso.

### 6. Gestión de cambios y gestión de puntos de cambio
* Al realizar cambios en las especificaciones del producto, el proceso, los equipos o los materiales y componentes, se evalúan los impactos de dichos cambios en las etapas de desarrollo, diseño, fabricación y servicio postventa, y se toman medidas preventivas con anticipación.
* Construcción del sistema y cumplimiento
* Verificación de la existencia de efectos secundarios
* Control de cambios en los proveedores externos
* Difusión de los detalles del cambio
*  Se trata de detectar y monitorear los puntos de cambio de manera temprana y tomar las medidas adecuadas. Un punto de cambio se refiere a situaciones en las que hay cambios en los factores o características, tales como fallos, interrupciones en el trabajo, cambios de operarios, o alteraciones ambientales externas como tormentas o terremotos.

### 7. Propósito, importancia y concepto de la inspección (conformidad y no conformidad)
* Evaluación de conformidad mediante observación y juicio, acompañada de mediciones adecuadas, pruebas o calibración de medidores.

### 8. Tipos y métodos de inspección
* La inspección se clasifica según la etapa en la que se realice, el método y la naturaleza.

### 9. Fundamentos de las normas
* La medición es el proceso de pensar y aplicar métodos y medios de cuantificar un objeto con un propósito específico, y utilizar los resultados para alcanzar el objetivo inicial. Un instrumento de medición es un término general que abarca aparatos, medidores,estándares, entre otros.

### 10. Gestión de medición
* Gestionar el sistema de actividades de medición.
* La calibración es un conjunto de actividades destinadas a estimar la relación entre el valor indicado por un instrumento o sistema de medición, o el valor representado por un patrón de medida o material de referencia, y el valor que se realiza mediante un estándar.
* Es importante la inspección y calibración de los instrumentos de medición.

### 11. Evaluación de los errores de medición
* Valor medido = Valor verdadero + Error
* Sesgo, Variabilidad, Exactitud (precisión global), Veracidad (exactitud), Precisión (exactitud)

### 12. Pruebas sensoriales / Calidad sensorial
* Se refiere a la evaluación sensorial a través de los cinco sentidos humanos, en la cual las características sensoriales se investigan mediante los órganos sensoriales de la persona y se evalúan en base a ello.
* Evaluación global, Evaluación absoluta,
 Evaluación relativa
* La calidad sensorial no solo se refiere a los cinco sentidos humanos, sino también a cómo se perciben aspectos como las emociones, estado de ánimo, grado de agrado, preferencias, confort, facilidad de uso, y la riqueza de la vida.
