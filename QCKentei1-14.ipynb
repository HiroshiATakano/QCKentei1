# Garantía de calidad

Para garantizar la calidad a los clientes, es requiere que la organización lleve a cabo sus actividades de forma sistemática.

## 14-01 Garantía de calidad

### 1. Garantía de resultados y garantía por proceso

**Garantía por proceso**: No se trata solo de garantizar los resultados mediante la inspección, sino de centrarse en el proceso (el trabajo) y gestionarlo y asegurar la calidad el mismo.
Esto también se expresa como que **la calidad se construye en el proceso**.

* Variabilidad del sistema de factores

### 2. Garantía y compensación

* **Garantía**: acciones enfocadas antes de que ocurra el problema.
  * Asegurar al cliente que todo está bien.
  * Garantizar el contenido y el rendimiento del producto.

* **Compensación**: acciones posteriores, principalmente de índole económica, después de que ocurra el problema
  * Indemnizar por los daños causados por el defecto.


### 3. Diagrama del sistema de garantía de calidad

**Diagrama del sistema de garantía de calidad**: es la asignación de las tareas relacionadas con la garantía de calidad en cada paso, desde el deseó y desarrollo del producto, la fabricación, la inspección, el envío, la venta, el servicio postventa, hasta el manejo de quejas.
El sistema de garantía de calidad se muestra generalmente en un **diagrama de flujo**.

A continuación se presentan los elementos que se deben incluir en el diagrama del sistema de aseguramiento de la calidad para que las actividades sean coherentes en todos los departamentos.

1. La **conexión de los procesos** en una secuencia cronológica que permite el **ciclo PDCA**.

2. **Criterios de evaluación**, **órganos de reunión**, **estándares y formularios** para avanzar en cada etapa.

3. La **ubicación de los roles** y **responsabilidades de cada departamento** en cada etapa.

* Roles de cada departamento. Ubicación de la responsabilidad


### 4. Despliegue de la función de calidad (QFD)

**QFD** (Quality Function Deployment):. es una metodología que utiliza diversas transformaciones y despliegues para lograr los objetivos de calidad del producto.

La **Despligue de la Función de Calidad** (QFD) organiza una serie de relaciones en múltiples **tablas bidimensionales**, desde la comprensión de los requisitos del cliente para el producto y la determinación de la calidad de diseño a lograr, hasta la calidad de las piezas que componen el producto y los elementos de control del proceso de fabricación.
De ellas, la tabla bidimensional que relaciona los **requisitos de calidad del cliente** con las **características de calidad del producto** se llama **Tabla de Calidad**.


### 5. DR y predicción de problemas FMEA FTA

En la aseguramiento de la calidad, se utilizan métodos como **DR**, **FMEA**, y **FTA**.


### 6. Proceso de garantía de calidad y red de aseguramiento

**Red de aseguramiento (QA network)**: es un diagrama que resume en qué proceso se deben realizar la **prevención de ocurrencia** y la  **prevención de salida**, a partir de la correspondencia entre los defectos y los procesos.

Los principales beneficios de crear una red de aseguramiento son:

1. Se vuelven claros los siguientes puntos: la importancia de los defectos, el nivel de aseguramiento, el **grado actual de aseguramiento** y el **estado de las contramedidas**.

2. Se pueden identificar los procesos criticos para prevenir defectos, lo que permite la **prevención en la fuente** y la **contención en el origen**.

3. Se pueden identificar los **procesos** en los que **se deben tomar acciones** para mejorar el nivel de aseguramiento general.

4. Al visualizar el estado del aseguramiento, se puede lograr un **entendimiento común** y la **transferencia de conocimientos y habilidades** entre las partes interesadas.





### 7. Garantía de calidad a lo largo de todo el ciclo de vida del producto

**Grantía de calidad a lo largo de todo el ciclo de vida del producto**: es la idea de qarantizar la calidad no solo en el momento de la venta del producto, sino también durante su uso e incluso hasta su eliminación.

Uno de esos métodos el el **Análisis del Ciclo de Vida** (LCA: Life Cycle Assessment), que es una técnica para **evaluar cuantitativamente el impacto ambiental** de un producto a lo largo de todo su ciclo de vida: desde la concepción y el desarrollo, pasando por la producción, el uso la reutilización, el reciclaje y la eliminación.


### 8. Seguridad del producto, consideraciones ambientales y responsabilidad por productos

**La seguridad del producto**: se refiere a la creación de productos seguros para garantizar la seguridad del cliente durante el uso del producto.

**La consideración ambiental** también se requiere en las normas relacionadas con las especificaciones del producto, así como en los métodos de prueba y evaluación correspondientes.

**La responsabilidad por productos defectuosos** (PL: Product Liability) es la responsabilidad de indemnización que debe asumir el fabricante por los daños personales o materiales causados por un defecto en el producto.

En Japón, la **Ley de Responsabilidad por Productos Defectuosos (PL)**, que establece la responsabilidad sin culpa para los fabricantes, se implementó a partir de julio de 1995.

La **responsabilidad sin culpa** es la obligación que tiene un fabricante u otro responsable de indeminizar por los daños, independientemente de si hubo negligencia por su parte.

La **Ley de Responsabilidad por Productos Defectuosos (PL)** se refiere a un producto que carece de la seguridad que normalmente debería tener y se puede clasificar en tres tipos:

1. **Defecto de diseño**: Un diseño defectuoso que carece de la seguridad que debería tener, ya que el problema reside en el propio diseño.

2. **Defecto de fabricación**: Un defecto de fabricación que carece de la seguridad que debería tener, ya que el producto no pudo ser fabricado según el diseño.

3. **Defecto en las instrucciones o advertencia**: Un defecto que consiste en no proporcionar la información adecuada sobre los peligros del producto, o en no dar las instrucciones o advertencias necesarías.

Los fabricantes dan gran importancia a las actividades de **Prevención de responsabilidad por productos defectuosos** (PLP: Product Liability Prevention) y **Defensa contra la responsabilidad por productos defectuosos** (PLD: Product Liability Defence) para evitar que surjan problemas de responsabilidad por sus productos.



### 9. Gestión de la fluidez inicial

**Gestión de la fluedez inicial**: Se trata de realizar una **gestión especial**, diferenciada de los productos anteriores, para la detección temprana y tratamiento rápido de los problemas relacionados con los defectos de calidad que no se detectaron en las etapas de diseño y prototipo, los cuales se hacen evidentes al **inicio de la producción de un nuevo producto** o **tras cambios significativos en el proceso de fabricación**, con el objetivo de estabilizar la calidad lo antes posible.


### 10. Manejo de problemas en el mercado, quejas y su resolución

Se pueden presentar quejas de los clientes sobre la calidad de los productos.
Afrontar las quejas es lo que se conoce como **respuesta a problemas del mercado** y es un asunto de suma importancia para los productores.

Una **queja** es una expresión de insatisfacción hacía una organización respecto a un producto, servicio o el proceso de gestión de quejas, en la que se espera explícita o implícitamente una respuesta o resolución.

El **manejo de quejas** es refiere a la investigación de las causas de la queja (utilizando métodos como el **análisis por qué-por qué**), la prevención de la reincidencia de quejas similares (incluyendo las de otros clientes), y la recopilación, el análisis y la utilización de la información de las quejas.

## 14-02 Garantía del proceso

### 1. Instrucción de trabajo (IT)

**Documento de estándar de trabajo**: Documento escrito del estándar de trabajo.

**Estándar de trabajo**: Estándar que indica el propósito del trabajo, las condiciones de trabajo (materiales utilizados, equipos y herramientas, entorno de trabajo, etc.), el método de trabajo (incluida la seguridad) y el método de verificación de los resultados del trabajo (autocontrol de calidad y cantidad, entre otros).


### 2. Concepto de proceso (o etapa del proceso)

**Proceso**: Es un conjunto de actividades interrelacionadas o que interactúan entre sí para convertir una entrada en una salida.


### 3. Diagrama de proceso QC y Diagrama de flujo

**Diagrama de proceso QC**: Es una representación en **diagrama de flujo** del proceso desde los materiales y componentes del producto hasta su envio como producoto terminado, en la que se clarifican los **elementos de control** y los **métodos de control** en **cada etapa del proceso**.



### 4. Concepto de anormalidades en el proceso, su detección y acciones correctivas

**Anomalía en el proceso** que el proceso no esté en estado de control y que deje de estar en estado estable debido a causas no despreciables.
Esto es diferente de una **no conformidad** (defecto), que ocurre cuando las características de un producto no cumplen con las especificaciones.

Para detectar una anomalía en el proceso, es fundamental el uso de **gráficos de control** y **sensibilidad de los operadores**.

Cuando se detecta una anomalía en el proceso, es recomendable emitir un **informe de anomalía del proceso** para responder con rapidez.




### 5. Investigación de la capacidad del proceso y análisis del proceso

La **capacidad del proceso** se refiere a la capacidad del proceso en relación con la calidad.
La investigación de la capacidad del proceso consiste en tomar muestras del proceso, medir las características de calidad y estimar la capacidad del proceso.

El **análisis del proceso** es una actividad que busca esclarecer la relación entre las características y los factores a partir de los datos recopilados en el proceso.
A menudo se utilizan métodos estadísticos para mantener y mejorar el proceso.



### 6. Gestión de cambios y gestión de puntos de cambio

**Gestión de cambios**: Al realizar cambios en las especificaciones del producto, el proceso, los equipos o los materiales y componentes, se evalúan los impactos de dichos cambios en las etapas de desarrollo, diseño, fabricación y servicio postventa, y se toman medidas preventivas con anticipación.


1. Construcción del sistema y cumplimiento
2.  Verificación de la existencia de efectos secundarios
3. Control de cambios en los proveedores externos
4. Difusión de los detalles del cambio

**Gestión de puntos de cambio**:  Se trata de detectar y monitorear los puntos de cambio de manera temprana y tomar las medidas adecuadas.
Un punto de cambio se refiere a situaciones en las que hay cambios en los factores o características, tales como fallos, interrupciones en el trabajo, cambios de operarios, o alteraciones ambientales extremas como tormentas o terremotos.




### 7. Propósito, importancia y concepto de la inspección (conformidad y no conformidad)

**Evaluación** de conformidad mediante observación y juicio, acompañada de mediciones adecuadas, pruebas o calibración de medidores.

El objetivo de la inspección es determinar si un producto individual es **conforme** o **no conforme** (defectuoso/producto defecuoso), o si un lote es **aprobado** o **no aprobado**, para evitar que los productos no conformes pasen a la siguiente estapa del proceso.

El objetivo también es transmitir la información obtenida de la inspección a las partes interesadas para tomar medidas de prevención de reincidencia y de prevención en las etapas anteriores del proceso.


### 8. Tipos y métodos de inspección

La inspección se clasifica según la etapa en la que se realice, el método y la naturaleza.

* Clasificación según la etapa en la que se realiza la inspección:
  * Inspección de recepción (inspección de compra)
  * Inspección entre procesos (inspección intermedia, inspección en proceso)
  * Inspección final
  * Inspección de envío
  * Autoinspección

* Clasificación según el método de inspección
  * Inspección al 100% (o inspección total)
  * Muestreo de inspección (o inspección muestreo)
  * Inspección indirecta
  * Inspección sin pruebas (o inspección no destructiva, si es el caso)

* Clasificación según la naturaleza de la inspección:
  * Inspección destructiva
  * Inspección no destructiva

* Clasificación según el lugar donde se realiza la inspección:
  * Inspección en puesto fijo
  * Inspección de patrulla



### 9. Fundamentos de las normas

La **medición** es el proceso de pensar y aplicar métodos y medios para cuantificar un objeto con un propósito específico, y utilizar los resultados para alcanzar el objetivo inicial.

Un **instrumento** de medición es un término general que abarca aparatos, medidores,estándares, entre otros.



### 10. Gestión de medición

**Gestión de medición**: Gestionar el sistema de actividades de medición.

La **calibración** es un conjunto de actividades destinadas a estimar la relación entre el valor indicado por un instrumento o sistema de medición, o el valor representado por un patrón de medida o material de referencia, y el vlor que se realiza mediante un estándar.

Es importante la inspección y calibración de los instrumentos de medición.



### 11. Evaluación de los errores de medición

Como la medición conlleva un error, es necesario analizar el error de medición para obtener mediciones precisas.

* Valor medido = Valor verdadero + Error

* Sesgo, Variabilidad, Exactitud (precisión global), Veracidad (exactitud), Precisión (exactitud)

El **Análisis del Sistema de Medición** (MSA: Measurement Systems Analysis), utilizado principalmente en la industria automotriz.

El **Gauge R&R** (Gage Repeatability and Reproducibility), una técnica de análisis representativa, es un método para evaluar la repetibilidad y reproducibilidad de las mediciones.
